```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# spacy for lemmatization
import spacy
```

```python
#Preparing StopWords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use','said',',','I','!','.','$','product','receive'])

#Obtaining the lemmatization data by score puntuation
nlp = spacy.load('en_core_web_sm')
global_lemm= []
for i in range(len(data)):
  doc = nlp(data[i])
  tokens= []
  for token in doc:
    tokens.append(token)
  lemm_text = " ".join([token.lemma_ for token in tokens])
  global_lemm.append(lemm_text)
```

```python
#Making the word Cloud by score
from wordcloud import WordCloud
for i in range(5):
  wordcloud = WordCloud(stopwords = stop_words,width= 500, height= 300 ,max_font_size=80, max_words=30, background_color="white").generate(global_lemm[i])
  # Display the generated image:
  # the matplotlib way:
  title = "Score: " + str(i+1)
  plt.title(title)
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.show()
```
